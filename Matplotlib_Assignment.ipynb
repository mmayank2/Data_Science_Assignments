Q2: What is a scatter plot? Use the following code to generate data for x and y. <br>Using this generated data
plot a scatter plot.<br>
import numpy as np<br>
np.random.seed(3)<br>
x = 3 + np.random.normal(0, 2, 50) <br>
y = 3 + np.random.normal(0, 2, len(x)) <br>
Note: Also add title, xlabel, and ylabel to the plot.</font>

<font color="green">Answer:-</font><br>
 A scatter plot is a type of data visualization used to display the relationship between two continuous variables. It is particularly useful when you want to understand the pattern or correlation between two numerical variables and identify any potential trends, clusters, or outliers in the data.

```python
import numpy as np
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))
```

```python
x,y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y)
plt.title("Relationship between x and y")
plt.xlabel("Values of x")
plt.ylabel("Values of y")
plt.show()
```

<font color="red">Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.</font>

<font color="green">Answer:-</font>
The suplot() function is used to create multiple plots within the same figure.it allows you to divide the figure into a grid of rows and columns and specify the position of each plot within that grid.

```python
plt.figure(figsize=(16,9))
x=np.array([0,1,2,3,4,5])
y1=np.array([0,100,200,300,400,500])
y2=np.array([50, 20, 40, 20, 60, 70])
y3=np.array([10, 20, 30, 40, 50, 60])
y4=np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x,y1,color="g",linestyle='dashed')
plt.title("Line plot")
plt.legend(['x1'])
plt.xlabel("This is x variable data")
plt.ylabel("This is y1 variable data")


plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Line plot")
plt.legend(["x,y2"])
plt.xlabel("This is x variable data")
plt.ylabel("This is y2 variable data")

plt.subplot(2,2,3)
plt.title("Line plot")
plt.plot(x,y3)
plt.legend(['x,y3'])
plt.xlabel("This is x variable data")
plt.ylabel("This is y3 variable data")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("Line plot")
plt.legend(["x,y4"],loc="upper left")
plt.xlabel("This is x variable data")
plt.ylabel("This is y4 variable data")
plt.show()
```

<font color="red">Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.</font>

<font color="green">Answer :-</font>
Bar plots are used to visualize categorical data and compare the values of different categories.

```python
company=np.array(["Apple", "Microsoft", "Google", "AMD"])
profit=np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(9,6))
plt.barh(company,profit,color=['r','y','m','k'])
plt.title("Company-profit")
plt.xlabel("Company Name")
plt.ylabel("Profit")
plt.show()
```

box1 = np.random.normal(100, 10, 200)<br>
box2 = np.random.normal(90, 20, 200)
</font>

<font color="green">Answer:-</font>
Box plot is a data visualization technique used to summarize the distribution of a continuous variable. It provides a compact representation of the key statistical measures of the data, such as the median, quartiles, and potential outliers.

Box plot is used because it provide concise and informative way to summarize and visualize the distribution of data like Visualizing Data Distribution,Identifying outlier, Comparing Distributions and etc.

```python
box1=np.random.normal(100,10,200)
box2=np.random.normal(90,20,200)
box1,box2
```

```python
box_colors=['pink','green']
plt.boxplot([box1,box2],labels=['Box1','Box2'],showmeans=True,showcaps=True,showbox=True,showfliers=True,patch_artist=True)
plt.show()
```
