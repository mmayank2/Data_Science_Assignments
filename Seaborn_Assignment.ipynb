<font color="red">Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.</font>

<font color="green">Answer :-</font>
Seaborn is a powerful data visualization library built on top of matplotlib. It provides a high level interface to create various types of statistical plots.
1. Bar Plot :-Bar plots are used to show the distribution of a categorical variable and compare the values of different categories.they are helpful for visualizing frequencies,counts or summary statistics for different groups.

2. Histogram :- Histograms are used to visualize the distribution of a continuous variable. They provide insights into data's shape,central tendency and spread.

3. Scatter plot :- Scatter plots are used to visualize the relationship between two continuous variables. They help identify correlations, clusters or pattern between variables.it is also useful for finding outliers.

4. Box plot :- Box plots are used to summarize the distribution of a continuous variable across different categories.They provide insights into the median,Quartiles,spread and presence of outliers for each category.

5. Heatmap :-Heatmaps are used to visualize the correlation between multiple variables in a dataset. They display a matrix of colors, where each cell represents the strength and direction of the relationship between two variables. Heatmaps are helpful for identifying patterns and dependencies in complex datasets.

<font color="red">Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.</font>

<font color="green">Answer:-</font>

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset("fmri")
```

```python
df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(16,9))
sns.lineplot(x="timepoint",y="signal",data=df,hue="event",style="region")
plt.show()
```

<font color="red">Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.</font>

```python
titanic_df=sns.load_dataset("titanic")
titanic_df.head()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
sns.boxplot(x="pclass",y="age",data=titanic_df)
plt.grid()

plt.subplot(1,2,2)
sns.boxplot(x="pclass",y="fare",data=titanic_df)
```

<font color="red">Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.</font>

```python
diamonds_df = sns.load_dataset("diamonds")
diamonds_df.head()
```

<font color="red">Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
